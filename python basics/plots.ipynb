```python
import random
import numpy as np
import matplotlib.pyplot as plt
class Plot:
    global x_axis
    global y_axis
    global explode
    global labels
    x_axis=np.random.randint(10,90,8)
    y_axis=np.random.randint(10,90,8)
    explode=[0,0,0,0.3,0,0,0,0]
    labels=["swarnadeep","rahul","raj","ram","rohan","abhiraj","coco","kolkata"]
    def plot_1():
        print("x-axis : ",x_axis)
        print("y-axis : ",y_axis)
        plt.plot(x_axis,marker="o",mfc="blue",color="magenta")
        plt.plot(y_axis,marker="o",mfc="red",color="indigo")
        plt.grid(color="green")
        plt.xlabel("students")
        plt.ylabel("placement")
        plt.title("SIT")
        plt.show()
    def plot_2():
        plt.pie(x_axis,explode=explode,shadow=True,labels=labels)
        plt.legend(labels)
Plot.plot_1()
Plot.plot_2()
```
